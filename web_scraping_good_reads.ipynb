```python
import selenium
import pandas as pd
```

```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome('C:/Users/Administrator/Desktop/chromedriver/chromedriver.exe')
```

```python
good_reads_url = "https://www.goodreads.com/choiceawards/best-books-2017"
browser.get(good_reads_url)
```

```python
reads = browser.find_elements_by_css_selector('div.category.clearFix')
print(len(reads))
type(reads)
```

```python
## 1.	From this website, extract the following information in a data frame format

#a)	Name of the category
#b)	Book name
#c)	URL to the book cover image
#d)	URL to the book description (Clicking on the image, takes you to a new page. Identify this URL)
```

```python
good_reads_df = pd.DataFrame(columns=['Category', 'Image_URL', 'Book_name','URL'])

for i in reads:
    name = i.find_element_by_css_selector('h4')
    name = name.get_attribute('innerHTML')
    url1 = i.find_element_by_css_selector('img.winner')
    img_url = url1.get_attribute('src')
    book_name = i.find_element_by_css_selector('img.winner')
    book_name = book_name.get_attribute('alt')
    url = i.find_element_by_css_selector('a')
    url = url.get_attribute('href')
    curr_df = {'Category' : name, 'Image_URL':img_url, 'Book_name' : book_name,'URL' : url}
    good_reads_df = good_reads_df.append(curr_df, ignore_index = True)

    
```

```python
type()
```

```python
#2.	From this webpage for each categories, collect the following information
#a)	Author name
#b)	Short description about the book
#c)	Votes for the winner
#d)	Total votes for all the books under that category
#e)	% of votes for the winner
```

```python
import re
l = pd.DataFrame(columns=['Author Name','Short Description','Votes','Total Votes','% of votes'])
for url in good_reads_df['URL']:
    browser.get(url)
    auth_nm = browser.find_element_by_css_selector('span[itemprop="name"]')
    auth_nm = auth_nm.get_attribute('innerHTML')
    short = browser.find_element_by_css_selector('div#description')
    short = short.get_attribute('innerHTML')
    vote = browser.find_element_by_css_selector('span.greyText.gcaNumVotes')
    vote = vote.get_attribute('innerHTML')
    vote = ''.join(re.findall("\d+\,\d+", vote)).replace(',','')
    tvote = browser.find_element_by_css_selector('greyText.gcaNumVotes')
    tvote = tvote.get_attribute('innerHTML')
    tvote = ''.join(re.findall("\d+\,\d+", tvote)).replace(',','')
    per_vote = round(int(vote) / int(tvote) * 100,2)
    
    curr_df = {'Author Name' : auth_nm,
               'Short Description' : short,
               'Votes' : vote,
               'Total Votes' : tvote,
               '% of votes' : per_vote}
    l = l.append(curr_df,ignore_index=True)
l
```

```python
top_books_2017 = pd.concat([good_reads_df,l],axis =1,ignore_index = False)
```
