```python
#!pip install selenium
import pandas as pd
import numpy as np
import selenium
from selenium import webdriver
```

```python
browser = webdriver.Chrome('D:/chromedriver.exe')
```

```python
twitter_url = "https://twitter.com/hashtag/ModiFailsNationalSecurity?src=tren"
browser.get(twitter_url)
```

```python
while browser.find_element_by_tag_name('div'):
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    Divs=browser.find_element_by_tag_name('button').text
    
    if 'Back to top ↑' in Divs:
        print('end')
        break
    else:
        continue
```

```python
friends = browser.find_elements_by_css_selector('div.tweet.js-stream-tweet.js-actionable-tweet.js-profile-popup-actionable.dismissible-content.original-tweet.js-original-tweet')
len(friends)
```

```python
browser.find_element_by_tag_name
```

```python
namo_tweets = pd.DataFrame(columns = ['Name','Text', 'Date', 'Likes'])
for i in friends:
    Name = i.find_element_by_css_selector('strong.fullname.show-popup-with-id.u-textTruncate').get_attribute('innerHTML')
    Text = i.find_element_by_css_selector('p.TweetTextSize.js-tweet-text.tweet-text').get_attribute('innerHTML')
    Date = i.find_element_by_css_selector('span._timestamp.js-short-timestamp').get_attribute('innerHTML')
    likes = i.find_element_by_css_selector('span.ProfileTweet-actionCountForPresentation').get_attribute('innerHTML')
    dictio = {'Name' : Name,'Text' : Text, 'Date' : Date, 'Likes' : likes}
    namo_tweets = namo_tweets.append(dictio, ignore_index=True)
```

```python
name = []
for i in namo_tweets['Name']:
    n = i.split('<')[0]
    name.append(n)
```

```python
namo_tweets = namo_tweets.drop('Name', axis = 1)
```

```python
namo_tweets['Name'] = name
```
