```python
browser = webdriver.Chrome('C:/Users/Administrator/Desktop/chromedriver/chromedriver.exe')
```

```python
myntra_url = "https://www.myntra.com/under-599-store"
browser.get(myntra_url)
```

```python
products = browser.find_elements_by_css_selector('li.product-base')
len(products)
```

```python
df_products = pd.DataFrame(columns = ['Name','Description','Price','Discount','URL'])

for item in products:
    name = item.find_element_by_css_selector('div.product-brand')
    name = name.get_attribute('innerHTML')
    desc = item.find_element_by_css_selector('h4.product-product')
    desc = desc.get_attribute('innerHTML')
    #print(name, '|', desc)
    discount = item.find_element_by_css_selector('span.product-discountPercentage')
    discount = discount.get_attribute('innerHTML')
    price_parent = item.find_element_by_css_selector('div.product-price') # class for all span isn't same, so we are pinpointing div
    price_child = price_parent.find_element_by_css_selector('span')   ## Here span is in div, so 1st we find span in div
    url = item.find_element_by_css_selector('a')
    url = url.get_attribute('href')
    try:
        price= price_child.find_element_by_css_selector('span.product-discountedPrice')
        price = price.get_attribute('innerHTML')
    except:
         price = price_child.get_attribute('innerHTML')
    curr_row = {'Name' : name,'Description' : desc,'Price' : price,'Discount' : discount,'URL' : url}
    df_products = df_products.append(curr_row, ignore_index = True)
df_products
```
